### Pandas (the 🐍 Python 🐍 library)

> Pandas (styled as pandas) is a software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series. The name is derived from the term "panel data", an econometrics term for data sets that include observations over multiple time periods for the same individuals, as well as a play on the phrase "Python data analysis". (From Wikipedia)

##### Pandas is a powerful data science library and provides the DataFrame implementation Esri (_the **API for Python** team_) has extended to provide robust **spatial** data support--more on that 🔜.

#### Key takeaways:
- Pandas is a package providing fast, flexible, and expressive data structures designed to make working with “relational” or “labeled” data both easy and intuitive
- Data feels like excel spreadsheets


#### Why use Pandas
- Easy handling of missing data
- Size mutability
- Automatic and explicit data alignment
- Powerful, flexible *group-by* functionality

#### Other Reasons
- Intelligent label-based slicing
- Fancy Indexing and subsetting
- Intuitive merging and joining data sets
- Flexible reshaping and pivoting of data sets

### NumPy

> NumPy (pronounced /ˈnʌmpaɪ/ NUM-py) is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays. NumPy is open-source software and has many contributors. (From Wikipedia)

#### Key takeaways
- NumPy provides data structures (e.g high-performance arrays, multi-dimensional arrays) and functions that support managing (or generating/randomizing) large amounts of mathematical data.
