### Pandas (the 🐍 Python 🐍 library)

> Pandas (styled as `pandas`) is a software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series. The name is derived from the term "panel data", an econometrics term for data sets that include observations over multiple time periods for the same individuals, as well as a play on the phrase "Python data analysis". (From Wikipedia)

##### Pandas is a powerful data science library and provides the DataFrame implementation Esri (_the **API for Python** team_) has extended to provide robust **spatial** data support--more on that 🔜.

#### Key takeaways:
- Pandas is a package providing fast, flexible, and expressive data structures designed to make working with “relational” or “labeled” data both easy and intuitive
- Data feels like excel spreadsheets


#### Why use Pandas
- Easy handling of missing data
- Size mutability
- Automatic and explicit data alignment
- Powerful, flexible *group-by* functionality

#### Other Reasons
- Intelligent label-based slicing
- Fancy Indexing and subsetting
- Intuitive merging and joining data sets
- Flexible reshaping and pivoting of data sets

#### Shuffling Data

When data is loading in order it's a good practice to shuffle it up before sampling.
Here's one way to do it:

```python
df = (iris_df
          # Return a random sample of items from an axis of object.
          .sample(frac=1)
          # Reset the index of the DataFrame, and use the default one instead.
          # Disassociates the new sample from the original order
          .reset_index(drop=True)
     )
df.head()
```

###### Method chaining / "Fluent" API

Notice the calls above.  Each of the called methods return a `DataFrame`, enabling calling other methods for further analysis.  Then, you can set the final result as the updated DataFrame to a new variable (or the original, if desired).

### Now let's build an index using the date

- Date/Time values can be used to index a DataFrame
- Helpful for time series data

```python
dates_with_hour = [[12, 25,2025,10], [12,31,2025,11], [1,1,2026, 0]]
df = pd.DataFrame(dates_with_hour, columns=['month', 'day', 'year', 'hour'])
# create the index just with the date; drop the hour
df.index = pd.to_datetime(df[['month', 'day', 'year']])
```

```python
# now that we have a datetime index, let's get the value for last Christmas
df.loc[pd.to_datetime('2025-12-25')]
```

#### Categorical Data
- Similar to domain values in Feature Classes
- used when you have a fixed number of values in a Series

```python
import numpy as np
# create a DataFrame with 20 random integers between 0 and 100
df = pd.DataFrame({'value': np.random.randint(0, 100, 20)})
labels = [f"{i} - {i + 9}" for i in range(0, 100, 10)]
# Bin values into discrete intervals.
# Use cut when you need to segment and sort data values into bins. 
# This function is also useful for going from a continuous variable to a categorical variable.
# from pandas docs ^
df['group'] = pd.cut(df.value, range(0, 105, 10), right=False, labels=labels)
df.head()
```

```python
df.dtypes
```
